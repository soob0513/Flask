# DB 연동 기능을 모듈로 만들기

- 모듈 : 함수들을 모아논 패키지 -> 재사용, 유지보수가 용이

- %%writefile : 다음 줄의 내용들을 해당 파일로 저장

```python
%%writefile ./DB_Control.py
import cx_Oracle

# DB 접속 함수
# username : 아이디, password : 암호, dsn : 접속주소
def db_conn(username, password, dsn) :
    try :
        conn = cx_Oracle.connect(username, password, dsn)
        cur = conn.cursor()

        print("DB 접속")
        return conn, cur
    except cx_Oracle.DatabaseError as e :
        print(e)

# DB 연결 종료 함수
# cur : 현재 커서 객체, conn : 현재 연결 객체
def db_disconn(conn, cur) :
    cur.close()
    conn.close()
    print("DB 접속 종료")

# DB에 데이터를 저장하는 함수
# t_name : 테이블명, col_dic : 컬럼명 딕셔너리
def insertData(conn, cur, t_name, col_dic) :
    # 딕셔너리의 키 값을 받아서 query문을 만들어 준다
    # col_dic -> {"code":code, "name":name, "age":age}
    #   -> insert into arome_tbl values (:code,:name,:age)
    query = f"insert into {t_name} values ("

    # insert into arome_tbl values (:code,:name,:age,
    for key in col_dic.keys() :
        query += f":{key},"

    # insert into arome_tbl values (:code,:name,:age
    # 가장 오른쪽에 있는 컴머를 제거
    query = query.rstrip(",")

    # insert into arome_tbl values (:code,:name,:age)
    query += ")"

    try :
        cur.execute(query, col_dic)
        conn.commit()

        if cur.rowcount != 0 :
            print("저장완료")
        else :
            print("저장실패")
    except cx_Oracle.DatabaseError as e :
        print(e)

# DB에서 해당 테이블의 전체 데이터를 검색하는 함수
def searchAllData(cur, t_name) :
    query = f"select * from {t_name}"

    try :
        cur.execute(query)
        row = cur.fetchall()

        return row
    except cx_Oracle.DatabaseError as e :
        print(e)

# DB에서 해당 테이블의 하나의 데이터를 검색하는 함수
# s_key : 검색할 컬럼명, s_value : 검색할 컬럼의 값
def searchData(cur, t_name, s_key, s_value) :
    query = f"select * from {t_name} where {s_key}='{s_value}'"

    try :
        cur.execute(query)
        row = cur.fetchone()

        return row
    except cx_Oracle.DatabaseError as e :
        print(e)

def searchData2(cur, t_name, s_key1, s_value1, s_key2, s_value2) :
    query = f"select name from {t_name} where {s_key1}='{s_value1}' and {s_key2}='{s_value2}'"

    try :
        cur.execute(query)
        row = cur.fetchone()

        return row
    except cx_Oracle.DatabaseError as e :
        print(e)

# DB에서 해당 테이블의 데이터를 수정하는 함수
# s_key : 수정할 컬럼명 (primary key인 컬럼)
def updateData(conn, cur, t_name, s_key, s_value, col_dic) :
    # update arome_tbl set
    query = f"update {t_name} set "

    # update arome_tbl set name=:name,age=:age,
    for key in col_dic.keys() :
        query += f"{key}=:{key},"

    # update arome_tbl set name=:name,age=:age
    query = query.rstrip(",")

    # update arome_tbl set name=:name,age=:age where code='code값';
    query += f" where {s_key}='{s_value}'"

    try :
        cur.execute(query, col_dic)
        conn.commit()

        if cur.rowcount != 0 :
            print("변경 완료")
        else :
            print("변경 실패")
    except cx_Oracle.DatabaseError as e :
        print(e)

# DB에서 해당 테이블의 데이터를 삭제하는 함수
def deleteData(conn, cur, t_name, s_key, s_value) :
    query = f"delete from {t_name} where {s_key}='{s_value}'"

    try :
        cur.execute(query)
        conn.commit()

        if cur.rowcount != 0 :
            print("삭제 완료")
        else :
            print("삭제 실패")
    except cx_Oracle.DatabaseError as e:
        print(e)
```
